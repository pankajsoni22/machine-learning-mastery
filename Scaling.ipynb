```python
import matplotlib.pyplot as plt
import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale
import pandas as pd
```

```python
address = 'data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

### Scalling of data

#### normalized 

```python
mpg = cars[['mpg']]
mpg_matrix = mpg.values
scaled = preprocessing.MinMaxScaler(feature_range=(0,5))
mpg_scaled = scaled.fit_transform(mpg_matrix)
plt.plot(mpg_scaled)
```

#### standardize scaling with mean 0 and unit variance

```python
standardized_mpg = scale(mpg)
```

```python
plt.plot(standardized_mpg)
```

```python
plt.plot(mpg_scaled, label='min-max scaler')
plt.plot(standardized_mpg, label='standard scaling')
plt.legend()
```
