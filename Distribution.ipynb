## Stastical Data Distribution
* **Density Function**: 
<ol>
    <li>Distributions are often described in terms of their density or density functions.</li>
    <li>Density functions are functions that describe how the proportion of data or likelihood of the proportion of observations change over the range of the distribution.</li>
</ol>

> Two types of density functions are probability density functions and cumulative density functions.
<div style="padding-left: 2rem;">
        <b>Probability Density function</b>: calculates the probability of observing a given value.<br/>
        <b>Cumulative Density function</b>: calculates the probability of an observation equal or less than a value.
</div>

### Types of distribution
<ol>
    <li>Gaussian Distribution</li>
    <li>Student's t-Distribution</li>
    <li>Chi Square Distribution sigma</li>
</ol>

```python
import numpy as np
```

```python
data = np.arange(-5,5,0.01)
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm,chi2
```

```python
mean = 0
stdev = 1
pdf = norm.pdf(data,mean, stdev)
pdf
```

```python
plt.plot(pdf)
```

```python
cdf = norm.cdf(data)
```

```python
plt.plot(cdf)
```

```python
pdf = chi2.pdf(data,df=4)
plt.plot(pdf)
```

```python
cdf = chi2.cdf(data,df=1)
plt.plot(cdf)
```
